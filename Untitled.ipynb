```python
# Clean curb data to remove any features that are not proper line segments
def filter_valid_line_segments(geojson_data):
    valid_features = []
    removed_features = []
    
    for feature in geojson_data['features']:
        # Check if it's a LineString with at least 2 points
        if (feature['geometry']['type'] == 'LineString' and 
            isinstance(feature['geometry']['coordinates'], list) and 
            len(feature['geometry']['coordinates']) >= 2):
            valid_features.append(feature)
        else:
            removed_features.append(feature)
    
    # Create a new GeoJSON with only valid features
    filtered_data = {
        "type": "FeatureCollection",
        "features": valid_features
    }
    
    print(f"Original features: {len(geojson_data['features'])}")
    print(f"Valid line segments: {len(valid_features)}")
    print(f"Removed invalid features: {len(removed_features)}")
    
    if removed_features:
        print("\nSample of removed features:")
        for i, feature in enumerate(removed_features[:3]):  # Show up to 3 examples
            print(f"\nRemoved feature {i+1}:")
            print(f"  Geometry type: {feature['geometry']['type']}")
            print(f"  Properties: {feature['properties'].get('label')}")
            print(f"  Coordinates: {feature['geometry']['coordinates']}")
    
    return filtered_data

# Apply the filter to our curb data
filtered_curb_data = filter_valid_line_segments(curb_data)

# Optionally visualize the filtered data
plot_spatial_data(non_street_cleaning_signs, filtered_curb_data)
```

```python
# More robust function to plot our spatial data
def plot_spatial_data(signs, curbs, figsize=(12, 12)):
    fig, ax = plt.subplots(figsize=figsize)
    
    # Plot curb segments with careful error checking
    valid_segments = []
    skipped_segments = 0
    
    for i, curb in enumerate(curbs['features']):
        try:
            coords = curb['geometry']['coordinates']
            
            # Skip if not a LineString or if coords isn't a list
            if curb['geometry']['type'] != 'LineString' or not isinstance(coords, list):
                skipped_segments += 1
                continue
                
            # Skip if not enough points
            if len(coords) < 2:
                skipped_segments += 1
                continue
                
            # Make sure each coordinate is properly formatted as [x, y]
            valid_coords = []
            for point in coords:
                if isinstance(point, (list, tuple)) and len(point) == 2:
                    valid_coords.append(point)
            
            # Skip if we don't have at least 2 valid points
            if len(valid_coords) < 2:
                skipped_segments += 1
                continue
                
            # Add the segment with valid coordinates
            valid_segments.append(np.array(valid_coords))
            
        except Exception as e:
            print(f"Error processing segment {i}: {e}")
            skipped_segments += 1
    
    # Only create LineCollection if we have valid segments
    if valid_segments:
        lc = LineCollection(valid_segments, colors='blue', linewidths=1, alpha=0.7)
        ax.add_collection(lc)
    
    # Plot signs
    try:
        sign_x = []
        sign_y = []
        for sign in signs:
            coords = sign['geometry']['coordinates']
            if isinstance(coords, (list, tuple)) and len(coords) == 2:
                sign_x.append(coords[0])
                sign_y.append(coords[1])
        
        if sign_x and sign_y:
            ax.scatter(sign_x, sign_y, c='red', s=10, alpha=0.7)
    except Exception as e:
        print(f"Error plotting signs: {e}")
    
    # Set aspect ratio and limits
    ax.set_aspect('equal')
    
    # Determine plot limits from data
    if valid_segments:
        all_points = np.vstack(valid_segments)
        x_min, y_min = np.min(all_points, axis=0)
        x_max, y_max = np.max(all_points, axis=0)
        
        # Add some padding
        padding = 0.001  # about 100 meters in lat/lon units
        ax.set_xlim(x_min - padding, x_max + padding)
        ax.set_ylim(y_min - padding, y_max + padding)
    else:
        ax.autoscale_view()
    
    plt.title('Signs and Curb Segments')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
    # Create legend
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], color='blue', lw=2, label='Curb Segments'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='red', 
               markersize=8, label='Signs')
    ]
    ax.legend(handles=legend_elements)
    
    plt.tight_layout()
    plt.show()
    
    print(f"Plotted {len(valid_segments)} valid line segments")
    print(f"Skipped {skipped_segments} invalid segments")
    print(f"Plotted {len(sign_x)} valid signs")

# Make sure to use the filtered data
plot_spatial_data(non_street_cleaning_signs, filtered_curb_data)
```

```python
def distance(point1, point2):
    """Calculate Euclidean distance between two points."""
    return math.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

def point_to_segment_distance(point, segment_start, segment_end):
    """Calculate the minimum distance from a point to a line segment."""
    # Vector from segment start to end
    segment_vector = [
        segment_end[0] - segment_start[0],
        segment_end[1] - segment_start[1]
    ]
    
    # Vector from segment start to point
    point_vector = [
        point[0] - segment_start[0],
        point[1] - segment_start[1]
    ]
    
    # Length of the segment squared
    segment_length_squared = segment_vector[0]**2 + segment_vector[1]**2
    
    # Handle degenerate case of zero-length segment
    if segment_length_squared == 0:
        return distance(point, segment_start), segment_start
    
    # Calculate projection of point_vector onto segment_vector (t parameter)
    t = max(0, min(1, (point_vector[0]*segment_vector[0] + point_vector[1]*segment_vector[1]) / segment_length_squared))
    
    # Find nearest point on segment
    projection = [
        segment_start[0] + t * segment_vector[0],
        segment_start[1] + t * segment_vector[1]
    ]
    
    # Return the distance and the projected point
    return distance(point, projection), projection

def find_nearest_segment(point, segments):
    """Find the nearest line segment to a point."""
    min_distance = float('inf')
    nearest_segment = None
    nearest_projection = None
    
    for segment in segments['features']:
        try:
            # Skip if not a LineString
            if segment['geometry']['type'] != 'LineString':
                continue
                
            coords = segment['geometry']['coordinates']
            
            # Skip if not enough points
            if len(coords) < 2:
                continue
            
            # For each line segment, find minimum distance to the point
            segment_min_distance = float('inf')
            segment_projection = None
            
            for i in range(len(coords) - 1):
                try:
                    segment_distance, projection = point_to_segment_distance(
                        point, 
                        coords[i],
                        coords[i+1]
                    )
                    
                    if segment_distance < segment_min_distance:
                        segment_min_distance = segment_distance
                        segment_projection = projection
                except Exception as e:
                    print(f"Error processing segment: {e}")
                    print(f"Coordinates: {coords[i]} to {coords[i+1]}")
                    continue
            
            if segment_min_distance < min_distance:
                min_distance = segment_min_distance
                nearest_segment = segment
                nearest_projection = segment_projection
        
        except Exception as e:
            print(f"Error processing feature: {e}")
            continue
    
    return nearest_segment, nearest_projection, min_distance

# Test with one sign and make sure to use filtered_curb_data
test_sign = non_street_cleaning_signs[0]
test_point = test_sign['geometry']['coordinates']
nearest_segment, projection, dist = find_nearest_segment(test_point, filtered_curb_data)

print(f"Test point: {test_point}")
if nearest_segment:
    print(f"Nearest segment ID: {nearest_segment['properties'].get('label')}")
    print(f"Projected point: {projection}")
    print(f"Distance: {dist}")
else:
    print("No valid segment found!")
```

```python
# Create a mapping of segments to nearby signs
segment_to_signs = {}

# Process each sign
for sign in non_street_cleaning_signs:
    sign_point = sign['geometry']['coordinates']
    nearest_segment, projection, dist = find_nearest_segment(sign_point, curb_data)
    
    if nearest_segment:
        segment_id = nearest_segment['properties'].get('label')
        
        if segment_id not in segment_to_signs:
            segment_to_signs[segment_id] = []
        
        # Calculate distance from segment start to projection
        line_coords = nearest_segment['geometry']['coordinates']
        dist_from_start = 0
        
        # Find which segment the projection is on
        for i in range(len(line_coords) - 1):
            segment_start = line_coords[i]
            segment_end = line_coords[i+1]
            
            # Check if projection is on this segment
            segment_dist, proj = point_to_segment_distance(projection, segment_start, segment_end)
            
            if segment_dist < 1e-10:  # Very close to this segment
                # Add distance from start to this segment
                for j in range(i):
                    dist_from_start += distance(line_coords[j], line_coords[j+1])
                
                # Add distance along this segment
                dist_from_start += distance(segment_start, projection)
                break
        
        segment_to_signs[segment_id].append({
            'sign': sign,
            'distance_from_start': dist_from_start,
            'projection': projection
        })

# Count how many segments have signs
print(f"Segments with assigned signs: {len(segment_to_signs)}")
print(f"Total segments: {len(curb_data['features'])}")

# Look at a sample segment and its signs
if segment_to_signs:
    sample_segment_id = list(segment_to_signs.keys())[0]
    print(f"\nSample segment {sample_segment_id} has {len(segment_to_signs[sample_segment_id])} signs:")
    for i, sign_info in enumerate(segment_to_signs[sample_segment_id]):
        print(f"  Sign {i+1}: {sign_info['sign']['properties'].get('regulation_type')}, " 
              f"Distance from start: {sign_info['distance_from_start']:.2f}")
```

```python
# Step 3: Assign regulation type to each segment based on closest sign to start
segment_regulation = {}

for segment in curb_data['features']:
    segment_id = segment['properties'].get('label')
    
    if segment_id in segment_to_signs and segment_to_signs[segment_id]:
        # Sort signs by distance from segment start
        sorted_signs = sorted(segment_to_signs[segment_id], key=lambda x: x['distance_from_start'])
        closest_sign = sorted_signs[0]['sign']
        regulation_type = closest_sign['properties'].get('regulation_type', 'Unknown')
        segment_regulation[segment_id] = regulation_type
    else:
        segment_regulation[segment_id] = None

# Count initial assignments
assigned_count = sum(1 for reg in segment_regulation.values() if reg is not None)
print(f"Initial segment assignments: {assigned_count} out of {len(curb_data['features'])}")

# Create a DataFrame to better visualize the assignments
assignments_df = pd.DataFrame([
    {
        'segment_id': segment_id,
        'regulation_type': regulation_type if regulation_type else 'None'
    }
    for segment_id, regulation_type in segment_regulation.items()
])

# Count by regulation type
regulation_counts = assignments_df['regulation_type'].value_counts()
print("\nRegulation type counts:")
print(regulation_counts)
```

```python
def are_segments_adjacent(segment1, segment2):
    """Check if two line segments are adjacent (share an endpoint)."""
    # Skip if either segment doesn't have a LineString geometry
    if segment1['geometry']['type'] != 'LineString' or segment2['geometry']['type'] != 'LineString':
        return False
    
    try:
        line1 = segment1['geometry']['coordinates']
        line2 = segment2['geometry']['coordinates']
        
        # Make sure both lines have at least 2 points
        if len(line1) < 2 or len(line2) < 2:
            return False
        
        # Make sure all coordinates are properly formatted
        if not all(isinstance(p, (list, tuple)) and len(p) == 2 for p in line1):
            return False
        if not all(isinstance(p, (list, tuple)) and len(p) == 2 for p in line2):
            return False
            
        # Check if any endpoint of one segment matches an endpoint of the other
        start1, end1 = tuple(line1[0]), tuple(line1[-1])
        start2, end2 = tuple(line2[0]), tuple(line2[-1])
        
        # Using a small epsilon to account for floating-point imprecision
        epsilon = 1e-6
        
        return (distance(start1, start2) < epsilon or 
                distance(start1, end2) < epsilon or 
                distance(end1, start2) < epsilon or 
                distance(end1, end2) < epsilon)
    except Exception as e:
        # If any error occurs, the segments aren't adjacent
        return False

# Build an adjacency map using filtered_curb_data
adjacency_map = {}
for segment in filtered_curb_data['features']:
    segment_id = segment['properties'].get('label')
    if segment_id is None:
        # Skip segments without an ID
        continue
        
    adjacency_map[segment_id] = []
    
    for other_segment in filtered_curb_data['features']:
        other_id = other_segment['properties'].get('label')
        
        if other_id is None:
            # Skip segments without an ID
            continue
            
        if segment_id != other_id and are_segments_adjacent(segment, other_segment):
            adjacency_map[segment_id].append(other_id)

# Check adjacency map
print("Sample of adjacency map:")
sample_keys = list(adjacency_map.keys())[:5]  # First 5 keys
for key in sample_keys:
    print(f"{key}: {adjacency_map[key]}")

# Count segments with no adjacent segments
no_adjacent = sum(1 for adj_list in adjacency_map.values() if not adj_list)
print(f"\nSegments with no adjacent segments: {no_adjacent}")
```

```python
# Copy the original regulation assignments
propagated_regulations = segment_regulation.copy()

# Step 4 & 5: Propagate regulation types to unassigned segments
changed = True
iteration_count = 0
max_iterations = 100  # Prevent infinite loops

# Track changes in each iteration
assignment_history = [sum(1 for reg in propagated_regulations.values() if reg is not None)]

while changed and iteration_count < max_iterations:
    changed = False
    iteration_count += 1
    
    for segment_id, adjacent_ids in adjacency_map.items():
        if propagated_regulations[segment_id] is None and adjacent_ids:
            # Check all adjacent segments for a regulation type
            for adj_id in adjacent_ids:
                if propagated_regulations[adj_id] is not None:
                    propagated_regulations[segment_id] = propagated_regulations[adj_id]
                    changed = True
                    break
    
    # Record how many segments have assignments after this iteration
    assignment_history.append(sum(1 for reg in propagated_regulations.values() if reg is not None))

# Count final assignments
final_assigned_count = sum(1 for reg in propagated_regulations.values() if reg is not None)
print(f"Final segment assignments after {iteration_count} iterations: {final_assigned_count} out of {len(curb_data['features'])}")

# Plot the assignment progress
plt.figure(figsize=(10, 6))
plt.plot(range(len(assignment_history)), assignment_history, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Number of Assigned Segments')
plt.title('Regulation Type Propagation Progress')
plt.grid(True)
plt.show()

# Create a DataFrame of final assignments
final_df = pd.DataFrame([
    {
        'segment_id': segment_id,
        'original_regulation': segment_regulation[segment_id] if segment_regulation[segment_id] else 'None',
        'propagated_regulation': propagated_regulations[segment_id] if propagated_regulations[segment_id] else 'None',
        'was_propagated': (segment_regulation[segment_id] is None and propagated_regulations[segment_id] is not None)
    }
    for segment_id in segment_regulation.keys()
])

print("\nFinal regulation type counts:")
print(final_df['propagated_regulation'].value_counts())

print("\nNumber of propagated assignments:")
print(final_df['was_propagated'].value_counts())
```

```python
def assign_metered_parking_near_streets(curb_segments, all_signs, max_distance=10):
    """
    Assigns 'Metered Parking' to curb segments that:
    1. Do not have a regulation_type assigned yet
    2. Are within max_distance meters of a sign on Tyler Street or Harrison Avenue
    
    Parameters:
    -----------
    curb_segments : dict
        GeoJSON FeatureCollection of curb segments
    all_signs : list
        List of ALL sign features (including street cleaning signs)
    max_distance : float
        Maximum distance in meters to consider (default: 10 meters)
    
    Returns:
    --------
    dict
        Updated GeoJSON with assigned regulation types
    """
    # Constants for converting degrees to meters
    # At Boston's latitude (42.35N)
    METERS_PER_DEGREE_LAT = 111320  # 1 latitude = ~111.32 km
    METERS_PER_DEGREE_LON = 83000   # 1 longitude = ~83 km at this latitude
    
    # Convert max_distance from meters to degrees (approximate)
    max_dist_lat = max_distance / METERS_PER_DEGREE_LAT
    max_dist_lon = max_distance / METERS_PER_DEGREE_LON
    
    # Find Tyler Street and Harrison Avenue signs
    target_street_signs = []
    
    for sign in all_signs:
        # Check all properties that might contain street information
        properties = sign['properties']
        street_fields = [
            properties.get('street_field', ''),
            properties.get('locator_street_field', ''),
            properties.get('notes_field', '')
        ]
        
        # Look for target streets in any of the fields
        is_target_street = False
        for field in street_fields:
            field_str = str(field).upper()
            if ('TYLER' in field_str or 'HARRISON' in field_str 
                or 'HARRISON AVE' in field_str or 'HARRISON AVENUE' in field_str):
                is_target_street = True
                break
        
        if is_target_street:
            target_street_signs.append(sign)
    
    print(f"Found {len(target_street_signs)} signs on Tyler Street or Harrison Avenue")
    
    # If we didn't find any, try a more lenient approach
    if len(target_street_signs) == 0:
        print("No signs found with exact street names. Trying a more lenient search...")
        
        # Get all signs' coordinates
        all_signs_coords = {}
        for sign in all_signs:
            sign_coords = tuple(sign['geometry']['coordinates'])
            if sign_coords in all_signs_coords:
                all_signs_coords[sign_coords].append(sign)
            else:
                all_signs_coords[sign_coords] = [sign]
        
        # Print some sample coordinates to debug
        print(f"Sample sign coordinates (first 3):")
        for i, coords in enumerate(list(all_signs_coords.keys())[:3]):
            print(f"  {i+1}: {coords}")
        
        # Let's try using bounding boxes for known areas around Tyler St and Harrison Ave
        tyler_bounds = {
            'min_lon': -71.068, 'max_lon': -71.058,
            'min_lat': 42.348, 'max_lat': 42.352
        }
        
        harrison_bounds = {
            'min_lon': -71.068, 'max_lon': -71.058,
            'min_lat': 42.346, 'max_lat': 42.350
        }
        
        for sign in all_signs:
            coords = sign['geometry']['coordinates']
            
            # Check if in Tyler St area
            if (coords[0] >= tyler_bounds['min_lon'] and coords[0] <= tyler_bounds['max_lon'] and
                coords[1] >= tyler_bounds['min_lat'] and coords[1] <= tyler_bounds['max_lat']):
                target_street_signs.append(sign)
                
            # Check if in Harrison Ave area
            elif (coords[0] >= harrison_bounds['min_lon'] and coords[0] <= harrison_bounds['max_lon'] and
                  coords[1] >= harrison_bounds['min_lat'] and coords[1] <= harrison_bounds['max_lat']):
                target_street_signs.append(sign)
        
        print(f"Found {len(target_street_signs)} signs in Tyler St or Harrison Ave bounding boxes")
    
    # Count curb segments without regulation type
    unassigned_segments = []
    for segment in curb_segments['features']:
        if (not segment['properties'].get('regulation_type') or 
            segment['properties'].get('regulation_type') == 'Unknown'):
            unassigned_segments.append(segment)
    
    print(f"Found {len(unassigned_segments)} curb segments without regulation type")
    
    # Function to check if a point is near any part of a line segment
    def is_point_near_segment(point, segment_coords, max_dist_lat, max_dist_lon):
        for i in range(len(segment_coords) - 1):
            # Simple bounding box check first (faster)
            min_lon = min(segment_coords[i][0], segment_coords[i+1][0]) - max_dist_lon
            max_lon = max(segment_coords[i][0], segment_coords[i+1][0]) + max_dist_lon
            min_lat = min(segment_coords[i][1], segment_coords[i+1][1]) - max_dist_lat
            max_lat = max(segment_coords[i][1], segment_coords[i+1][1]) + max_dist_lat
            
            if (point[0] >= min_lon and point[0] <= max_lon and 
                point[1] >= min_lat and point[1] <= max_lat):
                # More accurate distance check
                dist, _ = point_to_segment_distance(
                    point, 
                    segment_coords[i],
                    segment_coords[i+1]
                )
                
                # Convert approximated distance to meters
                dist_meters = dist * (METERS_PER_DEGREE_LAT + METERS_PER_DEGREE_LON) / 2
                
                if dist_meters <= max_distance:
                    return True
        
        return False
    
    # Function to check if a segment is near a sign
    def is_segment_near_sign(segment_coords, sign_point, max_dist_lat, max_dist_lon):
        return is_point_near_segment(sign_point, segment_coords, max_dist_lat, max_dist_lon)
    
    # Track how many segments we update
    updated_count = 0
    
    # Process each unassigned curb segment
    for segment in unassigned_segments:
        # Skip if not a LineString
        if segment['geometry']['type'] != 'LineString':
            continue
            
        segment_coords = segment['geometry']['coordinates']
        
        # Check if this segment is near any target street sign
        is_near_target_sign = False
        
        for sign in target_street_signs:
            sign_point = sign['geometry']['coordinates']
            
            if is_segment_near_sign(segment_coords, sign_point, max_dist_lat, max_dist_lon):
                is_near_target_sign = True
                break
        
        # Assign "Metered Parking" if near a target street sign
        if is_near_target_sign:
            segment['properties']['regulation_type'] = 'Metered Parking'
            segment['properties']['assigned_by'] = 'street_proximity'
            updated_count += 1
    
    print(f"Assigned 'Metered Parking' to {updated_count} curb segments near Tyler St/Harrison Ave signs")
    
    # If we still didn't assign any, let's debug by printing coordinates
    if updated_count == 0:
        print("\nDebugging coordinate ranges:")
        
        # Check bounds of signs
        sign_lons = [sign['geometry']['coordinates'][0] for sign in target_street_signs]
        sign_lats = [sign['geometry']['coordinates'][1] for sign in target_street_signs]
        
        print(f"Target street signs longitude range: {min(sign_lons)} to {max(sign_lons)}")
        print(f"Target street signs latitude range: {min(sign_lats)} to {max(sign_lats)}")
        
        # Check bounds of unassigned segments
        segment_coords = []
        for segment in unassigned_segments:
            if segment['geometry']['type'] == 'LineString':
                for coord in segment['geometry']['coordinates']:
                    segment_coords.append(coord)
        
        segment_lons = [coord[0] for coord in segment_coords]
        segment_lats = [coord[1] for coord in segment_coords]
        
        print(f"Unassigned segments longitude range: {min(segment_lons)} to {max(segment_lons)}")
        print(f"Unassigned segments latitude range: {min(segment_lats)} to {max(segment_lats)}")
        
        # Try with drastically increased distance
        very_large_dist = 100  # 100 meters
        max_dist_lat_large = very_large_dist / METERS_PER_DEGREE_LAT
        max_dist_lon_large = very_large_dist / METERS_PER_DEGREE_LON
        
        test_updated = 0
        for segment in unassigned_segments[:10]:  # Check just the first 10
            if segment['geometry']['type'] != 'LineString':
                continue
                
            segment_coords = segment['geometry']['coordinates']
            for sign in target_street_signs[:10]:  # Check just the first 10 signs
                sign_point = sign['geometry']['coordinates']
                if is_segment_near_sign(segment_coords, sign_point, max_dist_lat_large, max_dist_lon_large):
                    test_updated += 1
                    break
        
        print(f"With distance = 100m, {test_updated}/10 test segments would be updated")
    
    return curb_segments

# Apply this function to the segments data with ALL signs
# (including street cleaning signs to help find the streets)
updated_curb_data = assign_metered_parking_near_streets(filtered_curb_data, signs_data['features'])

# Check the updated regulation type counts
regulation_counts = {}
for segment in updated_curb_data['features']:
    reg_type = segment['properties'].get('regulation_type', 'Unknown')
    if reg_type not in regulation_counts:
        regulation_counts[reg_type] = 0
    regulation_counts[reg_type] += 1

print("\nRegulation type counts after Tyler St/Harrison Ave assignment:")
for reg_type, count in sorted(regulation_counts.items()):
    print(f"  {reg_type}: {count} segments")
```

```python
# Function to plot our spatial data with coloring by regulation type
def plot_results(curbs, figsize=(12, 12)):
    fig, ax = plt.subplots(figsize=figsize)
    
    # Create a colormap for regulation types
    import matplotlib.colors as mcolors
    from matplotlib.cm import get_cmap
    
    # Get unique regulation types
    reg_types = set()
    for curb in curbs['features']:
        reg_type = curb['properties'].get('regulation_type', 'Unknown')
        reg_types.add(reg_type)
    
    # Create a color map
    cmap = get_cmap('tab20', len(reg_types))
    reg_type_to_color = {reg_type: cmap(i) for i, reg_type in enumerate(reg_types)}
    
    # Plot curb segments with colors by regulation
    for curb in curbs['features']:
        coords = curb['geometry']['coordinates']
        reg_type = curb['properties'].get('regulation_type', 'Unknown')
        color = reg_type_to_color[reg_type]
        
        # Create line segments
        x = [p[0] for p in coords]
        y = [p[1] for p in coords]
        ax.plot(x, y, color=color, linewidth=2, alpha=0.7)
    
    # Set aspect ratio
    ax.set_aspect('equal')
    ax.autoscale_view()
    
    plt.title('Curb Segments Colored by Regulation Type')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
    # Create legend
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], color=reg_type_to_color[reg_type], lw=4, label=reg_type)
        for reg_type in reg_types
    ]
    ax.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.tight_layout()
    plt.show()

# Plot the results
plot_results(output_data)
```
